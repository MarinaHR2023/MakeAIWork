Wat is de gemiddelde prijs voor ruwe olie over de jaren?

```python
# Zoek in de Pandas documentatie een geschikte functie
# Hint: https://pandas.pydata.org/docs/reference/series.html

# https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html?highlight=average

# Voer deze functie uit op het veld 'Crude Oil' en print het resultaat

avgCrOil = com_pr["Crude Oil"].mean()
display(avgCrOil)
```

Wanneer was de olieprijs op zijn hoogst? En op zijn laagst?

```python
# Print highest and lowest price
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html?highlight=max#pandas.DataFrame.max
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.min.html?highlight=min#pandas.DataFrame.min
maxCrOil= com_pr["Crude Oil"].max()
display(maxCrOil)

minCrOil= com_pr["Crude Oil"].min()
display(minCrOil)
```

```python
# Print het jaar van de hoogste en laagste prijs
# https://pandas.pydata.org/docs/reference/api/pandas.Series.idxmax.html
# Series.idxmax(axis=0, skipna=True, *args, **kwargs)[source]
# Return the row label of the maximum value.
indMaxCrOil = com_pr['Crude Oil'].idxmax()
display(indMaxCrOil)

display(com_pr['Year'][52])
display(com_pr['Year'][indMaxCrOil])
display(com_pr.loc[52, "Year"])

#https://pandas.pydata.org/docs/reference/api/pandas.Series.idxmin.html#
# pandas.Series.idxmin
# Series.idxmin(axis=0, skipna=True, *args, **kwargs)
# Return the row label of the minimum value.

indMinCrOil = com_pr['Crude Oil'].idxmin()
display(indMinCrOil)
display(com_pr['Year'][indMinCrOil])
```

Net als SQL kan Pandas ook queries uitvoeren en filteren.

Zoek alle jaren waarin de olieprijs tussen 50 en 80 dollar was.

```python
# Hint: filter jaren waar de prijs tussen 50 en 80 was

crOilOv50= com_pr[com_pr["Crude Oil"] > 50] 
#display(crOilOv50)
crOilUn80= com_pr[com_pr["Crude Oil"] < 80] 
#display(crOilUn80)

# https://pandas.pydata.org/docs/user_guide/merging.html
# result = pd.concat([df1, df4], axis=1, join="inner")
crOilBet80a50 = pd.concat([crOilUn80,crOilOv50], axis=1, join="inner")
display(crOilBet80a50)
display(crOilBet80a50['Year'])
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.query.html
# Omdat 'Crude Oil' uit 2 woorden bestaat moet je een speciaal soort aanhalingstekens gebruiken - zie bovenstaande link
crOilBetw80a50 = com_pr.query("`Crude Oil` > 50 & `Crude Oil` < 80")
#display(crOilBetw80a50)
display(crOilBetw80a50['Year'])
```

## Opdracht 3: samenstellen

Stel dat Nutella zou bestaan uit 50% cacao, 20% suiker, 5% mais en 25% kokosolie.

Verder is gegeven:

- voor deze grondstoffen is de prijs per kilogram
- een pot bevat 800 gram Nutella

Wat is dan de prijs voor een pot Nutella in elk jaar?

```python
# Voeg een kolom 'Nutella' toe aan het DataFrame:
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/05_add_columns.html#min-tut-05-columns
# Voorbeeld: air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882

# Bereken de prijs voor een pot van 800 gram over de jaren heen:
# 50% cacao = 0.5*0.8*['Cocoa']
# 20% suiker = 0.2*0.8*['Sugar']
# 5% mais = 0.05*0.8*['Maize']
# 25% kokosolie = 0.25 * 0.8* ['Coconut Oil']


com_pr['Nutella'] = ((com_pr['Cocoa']*0.5) + (com_pr['Sugar']*0.2)+ (com_pr['Maize']*0.05) + (com_pr['Coconut Oil']*0.25))*0.8
print(com_pr.head())
print('Duur spul, dat Nutella!')
```
